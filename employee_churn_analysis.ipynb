# Connect to Big Query

```python
#libraries that we need
from google.cloud import bigquery
from google.colab import auth

#authenticate
auth.authenticate_user()

#initialize the client for big query
project_id = "employee-churn-project-454810"
client=bigquery.Client(project=project_id, location="US")
```

```python
#get the dataset and the table
dataset_ref=client.dataset("employeedata", project=project_id)
dataset=client.get_dataset(dataset_ref)
table_ref=dataset.table("tbl_hr_data")
table=client.get_table(table_ref)
table.schema
```

```python
new_table_ref=dataset.table("tbl_new_employees")
new_table=client.get_table(new_table_ref)
new_table.schema
```

```python
#Convert to dataframe
df=client.list_rows(table=table).to_dataframe()
df.head()
```

```python
#Convert to dataframe
df2=client.list_rows(table=new_table).to_dataframe()
df2.head()
```

## Code and train model

```python
#get our model
from pycaret.classification import *
```

```python
df.info()
```

```python
#set up our model
setup(df, target="Quit_the_Company", session_id=123, ignore_features=["employee_id"], categorical_features=["Departments", "salary"])
```

```python
compare_models()
```

```python
rf_model=create_model("rf")
```

```python
final_df=predict_model(rf_model)
```

```python
final_df.head()
```

```python
final_df.info()
```

```python
new_predicition = predict_model(rf_model, data=df2)
```

```python
new_predicition.head()
```

```python
#write back to big query
new_predicition.to_gbq("employeedata.pilot_predictions",
                       project_id,
                       chunksize=None,
                       if_exists="replace")
```

```python
plot_model(rf_model,plot="feature")
```

```python
#create a feature table
rf_model.feature_importances_
```

```python
import pandas as pd
feature_table = pd.DataFrame(zip(rf_model.feature_names_in_, rf_model.feature_importances_),
                             columns=["features","importance"])
feature_table
```

```python
feature_table.to_gbq("employeedata.feature_table",project_id, chunksize=None, if_exists="replace")
```
